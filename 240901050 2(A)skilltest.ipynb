```python
import numpy as np
steps = np.array([5000, 6500, 7000, 8000, 9000, 7500, 10000, 8500, 9200, 11000])
print("\nTotal steps:", np.sum(steps))
print("Average steps per day:", np.mean(steps))
print("Maximum steps:", np.max(steps))
print("Minimum steps:", np.min(steps))
print("\nHighest activity days(>9000):")
for i in range (len(steps)):
    if steps[i] >9000:
       print("Day",i+1,":",steps[i])
trend = np.diff(steps)
print("\nDaily change in steps:", trend)
```
